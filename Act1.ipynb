```python
import pandas as pd
```

```python
name = '/content/data.tsv'

def read_csv(name):
  return pd.read_csv(name, sep='\t')

def quantity():
  return read_csv(name)['quantity'].sum()
  
print('El numero de pedidos es:', quantity())
```
